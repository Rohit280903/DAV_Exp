```python
import matplotlib.pyplot as plt
import numpy as np

# Matplotlib Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Matplotlib Line Plot')
plt.show()

# Matplotlib Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title('Matplotlib Scatter Plot')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Plotly Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))
fig.update_layout(title='Plotly Line Plot')
fig.show()

# Plotly Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers'))
fig.update_layout(title='Plotly Scatter Plot')
fig.show()
```

```python
# ggplot2 Line Plot
x <- seq(0, 10, length.out = 100)
y <- sin(x)
ggplot(data.frame(x, y), aes(x, y)) + geom_line() + ggtitle('ggplot2 Line Plot')

# ggplot2 Scatter Plot
data <- data.frame(x = rnorm(50), y = rnorm(50))
ggplot(data, aes(x, y)) + geom_point() + ggtitle('ggplot2 Scatter Plot')
```

```python
# Lattice Line Plot
x <- seq(0, 10, length.out = 100)
y <- sin(x)
xyplot(y ~ x, type = 'l', main = 'Lattice Line Plot')

# Lattice Scatter Plot
data <- data.frame(x = rnorm(50), y = rnorm(50))
xyplot(y ~ x, data = data, type = 'p', main = 'Lattice Scatter Plot')
```
